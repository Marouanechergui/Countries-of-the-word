#Remplacer les valeur manquante par la mediace de chaque colonne

```python
liste=['Net migration','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Crops (%)','Other (%)','Climate','Birthrate','Deathrate','Agriculture','Industry','Service']
for col in data.columns:
    if data[col].isnull().sum()!=0:
        if col in liste:
            data[col] =data.groupby('Region')[col].transform(lambda x: x.fillna(x.median()))
```

```python
#Entrainement des models:
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.set_index(['Region','Country'],inplace=True)
```

```python
data.head()
```

```python
X= data.drop('GDP ($ per capita)',axis=1).values
y = data['GDP ($ per capita)'].values
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.45, random_state=12)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
scaler=MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
y_pred= linreg.predict(X_test)
```

```python
r2 = r2_score(y_test,y_pred)
print(r2)
MSE = mean_squared_error(y_test,y_pred)
print(MSE)
MSA = mean_absolute_error(y_test,y_pred)
print(MSA)
```

Cross val

```python
from sklearn.model_selection import cross_val_score
print(np.mean(cross_val_score(linreg,X_train,y_train,cv=5)))
```

```python
print(linreg.coef_)
```

```python
names = data.drop('GDP ($ per capita)',axis=1).columns
fig,ax = plt.subplots(figsize=(7,7))
a = plt.plot(range(len(names)),linreg.coef_)
b=plt.xticks(range(len(names)),names,rotation=90)
```

Construction du ridge

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge( alpha= 0.3)
```

```python
ridge.fit(X_train,y_train)
```

```python
y_pred_ridge = ridge.predict(X_test)
```

```python
r2ridge = r2_score(y_test,y_pred_ridge)
print(r2)
MSE = mean_squared_error(y_test,y_pred_ridge)
print(MSE)
MSA = mean_absolute_error(y_test,y_pred_ridge)
print(MSA)
```

```python
print(np.mean(cross_val_score(ridge,X_train, y_train , cv = 6)))
```

```python
names = data.drop('GDP ($ per capita)',axis=1).columns
fig,ax = plt.subplots(figsize=(7,7))
a = plt.plot(range(len(names)),ridge.coef_)
b=plt.xticks(range(len(names)),names,rotation=90)
```

entrainement du model random forest

```python
r2random = r2_score(y_test,y_pred_forest)
print('R2 = ' ,r2)
MSE = mean_squared_error(y_test,y_pred_forest)
print('MSE = ',MSE)
MSA = mean_absolute_error(y_test,y_pred_forest)
print('MSA = ',MSA)
```

```python
rmse= np.sqrt(np.mean(-cross_val_score(forest_model, X_train, y_train,cv=5, scoring='neg_mean_squared_error')))
print('RMSE:%f' %(rmse))
r2_forest = np.mean(cross_val_score(forest_model,X_train,y_train, cv=5, scoring='r2'))

print('R2 score: %s' %'{:2}'.format(r2_forest))
```
